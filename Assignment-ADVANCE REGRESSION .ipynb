# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler

# Assuming you have training data x_train and y_train
# x_train, y_train = ...

# Standardize features (recommended for Ridge regression)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)

# Set up Ridge regression model
ridge = Ridge()

# Set up a parameter grid for alpha (the regularization parameter)
param_grid = {'alpha': [0.1, 1, 10, 100]}  # You can adjust the values as needed

# Use GridSearchCV to find the best alpha
grid_search = GridSearchCV(ridge, param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(x_train_scaled, y_train)

# Get the best alpha from the grid search
best_alpha = grid_search.best_params_['alpha']

# Train the Ridge regression model with the best alpha
ridge_model = Ridge(alpha=best_alpha)
ridge_model.fit(x_train_scaled, y_train)

# Display the coefficients of the Ridge model
print("Ridge Regression Coefficients:")
print(ridge_model.coef_)

# Display other relevant information
print(f"Best alpha: {best_alpha}")
print(f"Intercept: {ridge_model.intercept_}")
```

# Lasso Regression 

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler


# Standardize features (recommended for Lasso regression)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)

# Set up Lasso regression model
lasso = Lasso()

# Set up a parameter grid for alpha (the regularization parameter)
param_grid = {'alpha': [0.1, 1, 10, 100]}  # You can adjust the values as needed

# Use GridSearchCV to find the best alpha
grid_search = GridSearchCV(lasso, param_grid, scoring='neg_mean_squared_error', cv=5)
grid_search.fit(x_train_scaled, y_train)

# Get the best alpha from the grid search
best_alpha = grid_search.best_params_['alpha']

# Train the Lasso regression model with the best alpha
lasso_model = Lasso(alpha=best_alpha)
lasso_model.fit(x_train_scaled, y_train)

# Display the coefficients of the Lasso model
print("Lasso Regression Coefficients:")
print(lasso_model.coef_)

# Standardize features for the test set using the same scaler
x_test_scaled = scaler.transform(x_test)

# Calculate R^2 score for Lasso regression
lasso_r2 = lasso_model.score(x_test_scaled, y_test)

print(f"R^2 Score for Lasso Regression: {lasso_r2}")
```
